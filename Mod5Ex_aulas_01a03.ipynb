```python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
# 2) seu código aqui
sinasc.isnull().sum()
```

```python
# 3) seu código aqui
#criando uma nova tabela com o que precisamos
df = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']]
# recontando os nulos
df.isnull().sum()
```

```python
#vamos ver como esta a tabela:
df
```

```python
# 4) seu código aqui
#usando o subset para deletar as linhas com nulos em uma coluna especifica.
df = df.dropna(subset=['APGAR5'])
df.isnull().sum()
```

```python
# 5) seu código aqui
#vamos usar .fillna com dicionario para mostrar qual colunas queremos substituir.
df = df.fillna({'ESTCIVMAE': 9, 'CONSULTAS': 9})
```

```python
# 6) Seu código aqui
#vamos usar .fillna com dicionario para mostrar qual colunas queremos substituir.
df = df.fillna({'QTDFILVIVO': 0})
df.isnull().sum()
```

```python
# 7) seu código aqui
# na coluna GRAVIDEZ vamos retirar as linhas com NaN, ja que queremos analisar
#a relação entre partos com risco para o bebê e algumas condições como tempo de parto
df = df.dropna(subset=['GRAVIDEZ'])


# PAra a coluna ESCMAE também vamos tirar os Nan, coma  variabilidade metabólica entre as idades da criaça assumir
# qualquer valor aqui pode nos levar a uma conclusão errada, é melhor utilizar apenas os dados que temos certeza. 
df = df.dropna(subset=['ESCMAE'])

# Para a gestação vamos fazer o mesmo devido as grandes diferenças entre etapas da gestação, com isso vamos perder cerca de  
# 4,5% dos dados, mas nosso modelo será mais acurado 
df = df.dropna(subset=['GESTACAO'])

df.isnull().sum()
```

```python
# 8) seu código aqui

df.loc[df['APGAR5'] <= 3, 'asf_cat'] = 'asfixia severa'
df.loc[(df['APGAR5'] > 3) & (df['APGAR5'] <= 5),
       'asf_cat'] = 'asfixia moderada'
df.loc[(df['APGAR5'] > 5) & (df['APGAR5'] <= 7),
       'asf_cat'] = 'asfixia leve'
df.loc[(df['APGAR5'] > 7), 'asf_cat'] = 'normal'

```
